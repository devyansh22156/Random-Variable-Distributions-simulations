```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
import math
import textwrap
```

This block of code defines a function called aBernoulliExperiment, which simulates a Bernoulli experiment by generating a specified number of random binary outcomes (0 or 1) with a specified probability of success (i.e., the parameter bernoulliParam_p). The function returns a numpy array of size (numRepeatsOfExp, numTrialsInAnExp) containing the simulated outcomes.

```python
def aBernoulliExperiment(randomNumGen, bernoulliParam_p=0.2, numRepeatsOfExp=1, numTrialsInAnExp=20, profitLoss=[1,0]):
    observation = np.ones((numRepeatsOfExp, numTrialsInAnExp))
    repeatIndex = 0
    while repeatIndex < numRepeatsOfExp:
        observation[repeatIndex,:] = randomNumGen.choice([0,1], p=[1-bernoulliParam_p,bernoulliParam_p], size=numTrialsInAnExp)
        repeatIndex += 1
    return observation
```

The second block of code generates a plot using matplotlib that shows the outcomes of 20 trials of a coin with probability of heads 0.2. Specifically, the code uses the aBernoulliExperiment function to generate a single set of outcomes for 20 trials, and then plots those outcomes as a step function using the ax.step method. The where='post' argument specifies that the step function should take on the value of the previous point at each x-coordinate. The resulting plot shows the outcomes of each trial as a step function with 20 steps, where the height of each step corresponds to the outcome of that trial (0 or 1). The plot also includes a legend indicating which repeat the outcomes correspond to (in this case, there is only one repeat).

```python
fig, ax = plt.subplots()
for i in range(1):
    observation = aBernoulliExperiment(np.random.default_rng(), bernoulliParam_p=0.2, numRepeatsOfExp=1, numTrialsInAnExp=20, profitLoss=[1,0])
    print(observation)
    ax.step(range(1, 21), observation[0], where='post', label='Repeat ' + str(i+1))
ax.set_xlabel('Trial Number')
ax.set_ylabel('Outcome')
ax.set_ylim(-0.1, 1.1)
ax.set_title('Outcomes of 20 Trials of a Coin with Probability 0.2')
ax.legend(loc='best')
plt.show()
```

                                                        Question 2

Pick four values of $\alpha$ uniformly and randomly from the interval $(0, 100)$. For each $\alpha$, simulate a Binomial $(n,p)$, with $n$ chosen for the $\alpha$ to be as small as possible while providing a reasonable approximation (the PMFs/ histograms look visually similar) of the Poisson $(\alpha)$. Note that you must also simulate the Poisson RVs. Choose the number of experiements for all the RVs to be large enough to ensure that the resulting histograms are good approximations of the PMFs.

```python
from scipy.stats import norm
from scipy.special import comb
import matplotlib.pyplot as plt
import numpy as np
import math
import textwrap


def aBernoulliExperiment(randomNumGen, bernoulliParam_p=0.5, numRepeatsOfExp=1, numTrialsInAnExp=1, profitLoss=[1,0]):
    observation = np.ones((numRepeatsOfExp, numTrialsInAnExp))
    repeatIndex = 0
    while repeatIndex < numRepeatsOfExp:
        observation[repeatIndex,:] = randomNumGen.choice([0,1], p=[1-bernoulliParam_p,bernoulliParam_p], size=numTrialsInAnExp)
        repeatIndex += 1
    return observation

def aBinomialExperiment(randomNumGen, binomialParam_n=20, binomialParam_p=0.2, numberOfBinomialExperiments = 1, profitLoss_Bern = [1,0]):

    outcomesOfBernTrails = aBernoulliExperiment(randomNumGen, bernoulliParam_p=binomialParam_p, numRepeatsOfExp = numberOfBinomialExperiments, numTrialsInAnExp = binomialParam_n, profitLoss = profitLoss_Bern)

    outcomeOfBinomialExp = np.sum(outcomesOfBernTrails, axis = 1)

    return outcomeOfBinomialExp

alphas = np.random.uniform(0, 100, size=4)

ps = 0.5
ns = [int(alpha/ps) + 1 for alpha in alphas]

poissons = [np.random.poisson(alpha, size=10000) for alpha in alphas]

binomials = [aBinomialExperiment(np.random.default_rng(7381638), binomialParam_n=20, binomialParam_p=0.2, numberOfBinomialExperiments=10000) for n in ns]

fig, axs = plt.subplots(2, 4, figsize=(16, 8))

for i in range(4):
    axs[0, i].hist(poissons[i], bins=20, rwidth=0.6, color='blue', density=True)
    axs[0, i].set_title(f'Poisson({alphas[i]:.2f})')
    axs[1, i].hist(binomials[i], bins=20, rwidth=0.6, color='red', density=True)
    axs[1, i].set_title(f'Binomial({ns[i]}, {ps:.1f})')

print("Poissons: ")
print(poissons)
print("Binomials: ")
print(binomials)


plt.tight_layout()
plt.show()

```

The code generates four values of alpha, calculates the minimum value of n for each alpha, simulates Poisson distributions and Binomial distributions, and plots the histograms of the simulated data.

In the output graph, the top row shows the histograms of Poisson distributions with different alpha values, and the bottom row shows the histograms of Binomial distributions with different n values. The histograms are colored blue for Poisson distributions and red for Binomial distributions. The x-axis represents the number of occurrences of the event, and the y-axis represents the frequency of the occurrence.

We can observe that as the value of alpha increases, the distribution of Poisson becomes more spread out and skewed to the right. On the other hand, as the value of n increases, the distribution of Binomial becomes more symmetric and closer to the normal distribution.

Illustrate the law of large numbers and the central limit theorem for a sum of iid Poisson RVs. Note that the provided code (Binomial.ipynb) does the same for a sum of Bernoulli $(p)$ RVs.

Do the above for two values of $\alpha$ chosen uniformly and randomly from $(1, 20)$. For each value of $\alpha$ demonstrate the above over an increasing, four long, sequence of number of RVs being added. For example, in the provided code, we simulated $5,10,100,1000$ Bernoulli RVs being summed. Note that you may need a sequence different from the example. Figure out a sequence that shows convergence to a bell shaped curve (central limit theorem) and clustering of the histogram around the mean (law of large numbers).

```python
from scipy.stats import poisson, norm
import matplotlib.pyplot as plt
import numpy as np

def aPoissonExperiment(randomNumGen, poissonParam_lambda=1, numRepeatsOfExp=1, numRVsInAnExp=20):
    observation = np.ones((numRepeatsOfExp, numRVsInAnExp))
    repeatIndex = 0
    while repeatIndex < numRepeatsOfExp:
        observation[repeatIndex,:] = randomNumGen.poisson(poissonParam_lambda, size=numRVsInAnExp)
        repeatIndex += 1
    return observation

alphas = np.random.uniform(1, 20, size=2)

ns = [int(alpha) + 1 for alpha in alphas]

poissons = [aPoissonExperiment(np.random.default_rng(), poissonParam_lambda=alpha, numRepeatsOfExp=10000, numRVsInAnExp=n) for alpha, n in zip(alphas, ns)]

poisson_sums = [np.sum(poisson, axis=1) for poisson in poissons]

fig, axs = plt.subplots(2, 4, figsize=(16, 6))

for i in range(2):
    for j, n in enumerate([5, 10, 500, 1000]):
        subset = poisson_sums[i][:10000000*(j+1)]
        axs[i, j].hist(subset, bins=20, color='blue', rwidth=0.4, density=True)
        axs[i, j].set_title(f'n={n}, alpha={alphas[i]:.2f}')

        x = np.linspace(0, np.max(subset), 100)
        poisson_pmf = poisson.pmf(x, alphas[i]*n)
        axs[i, j].plot(x, poisson_pmf, label='Poisson')
        normal_pdf = norm.pdf(x, alphas[i]*n, np.sqrt(alphas[i]*n))
        axs[i, j].plot(x, normal_pdf, label='Normal')
        axs[i, j].legend()

        print(f"alpha={alphas[i]:.2f}, n={n}, Mean={np.mean(subset):.2f}, Std Dev={np.std(subset):.2f}")

plt.tight_layout()
plt.show()
```

This code simulates Poisson distributions with different values of lambda chosen randomly between (1, 20) and sample sizes n = [5, 10, 500, 1000], calculates the sum of Poisson random variables, and plots histograms of the results along with the expected Poisson and normal distributions.

The generated plots show that as the sample size increases, the Poisson distribution tends to resemble the normal distribution. This is due to the Central Limit Theorem, which states that as the sample size increases, the distribution of the sample means approaches a normal distribution regardless of the shape of the population distribution.

Overall, the plots demonstrate the relationship between Poisson and normal distributions, and the effect of sample size on the Bell shape of the distribution.
